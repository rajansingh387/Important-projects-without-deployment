```python
ratings_with_names= ratings.merge(books,on='ISBN')
```

```python
ratings_with_names.head()
```

```python
num_of_ratings= ratings_with_names.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_of_ratings.rename(columns={'Book-Rating':'num_of_ratings'},inplace= True)
num_of_ratings
```

```python
avg_of_ratings= ratings_with_names.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_of_ratings.rename(columns={'Book-Rating':'avg_of_ratings'},inplace= True)
avg_of_ratings
```

```python
popular_ratings= num_of_ratings.merge(avg_of_ratings,on='Book-Title')
```

```python
popular_ratings
```

```python
popular_df=popular_ratings[popular_ratings['num_of_ratings']>=250].sort_values('num_of_ratings',ascending= False).head(50)
```

```python
x= ratings_with_names.groupby('User-ID')['Book-Rating'].count()>200
experienced= x[x].index
```

```python
ratings_with_names
```

```python
filterd_rating =ratings_with_names[ratings_with_names['User-ID'].isin(experienced)]
```

```python
filterd_rating
```

```python
y=filterd_rating.groupby('Book-Title').count()['Book-Rating']>=10
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings= filterd_rating[filterd_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt= final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend_system(book_name):
  index= np.where(pt.index==book_name)[0][0]
  similar_books=  sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1], reverse= True)[1:11]
  for i in similar_books:
    print(pt.index[i[0]],f'\n similarity score is {i[1]}')
```

```python
recommend_system('White Oleander : A Novel')
```
